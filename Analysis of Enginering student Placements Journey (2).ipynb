## **Data Cleaning**

```python
# 1. Replace nulls in critical ID/Categorical columns with 'Unknown'
critical_categorical = ['Student ID', 'Gender', 'Branch', 'Placement Status']
df[critical_categorical] = df[critical_categorical].fillna('Unknown')
```

```python
# 2. Fill numeric columns with column mean
numeric_cols = [
    'Age', 'Average GPA', 'Backlogs', 'Attendance (%)', 'CTC (LPA)',
    'Sem1 GPA', 'Sem2 GPA', 'Sem3 GPA', 'Sem4 GPA',
    'Sem5 GPA', 'Sem6 GPA', 'Sem7 GPA', 'Sem8 GPA'
]
df[numeric_cols] = df[numeric_cols].apply(lambda col: col.fillna(col.mean()))

```

```python
# 3. Fill text columns with 'Unknown'
text_cols = ['Clubs', 'Skills', 'Internship Done']
df[text_cols] = df[text_cols].fillna('Unknown')
```

```python
# 4. Fill high-null columns with 'Unknown'
high_null_text_cols = ['Internship Domain', 'Placement Domain', 'Alumni Path']
df[high_null_text_cols] = df[high_null_text_cols].fillna('Unknown')
```

### **Questions and Answers**

 Q1: What is the total number of students in the dataset?

```python
#Total number of students
total_students = len(df)
total_students
```

 Q2: What are the different branches represented?

```python
unique_branches = df['Branch'].dropna().unique().tolist()
unique_branches
```

 Q3: What is the average GPA across all students?

```python
average_gpa = round(df['Average GPA'].mean(), 2)
average_gpa
```

 Q4: Which student has the highest and lowest GPA?

```python
top_student = df.loc[df['Average GPA'].idxmax(), ['Name', 'Average GPA']]
top_student 

bottom_student = df.loc[df['Average GPA'].idxmin(), ['Name', 'Average GPA']]
bottom_student
```

Q5: What is the gender distribution?

```python
gender_distribution = df['Gender'].value_counts().to_dict() 
gender_distribution
```

Q6: What is the average GPA by gender?

```python
avg_gpa_by_gender = df.groupby('Gender')['Average GPA'].mean().round(2).to_dict()
avg_gpa_by_gender
```

Q7: Which branch has the highest average GPA?

```python
avg_gpa_by_branch = df.groupby('Branch')['Average GPA'].mean().round(2).sort_values(ascending=False).to_dict()
avg_gpa_by_branch
```

 Q8: How many students have done internships?

```python
internship_participation = df['Internship Done'].value_counts().to_dict()
internship_participation
```

 Q9: What percentage of students got placed?

```python
placement_status_pct = df['Placement Status'].value_counts(normalize=True).mul(100).round(2).to_dict()
placement_status_pct 
```

Q10: What is the average CTC for placed students?

```python
avg_ctc_placed = round(df[df['Placement Status'] == 'Placed']['CTC (LPA)'].mean(), 2)
avg_ctc_placed 
```

 Q11: What is the average number of backlogs per student?

```python
avg_backlogs = round(df['Backlogs'].mean(), 2)
avg_backlogs
```

Q12: Which is the most popular club?

```python
most_popular_club = df['Clubs'].value_counts().idxmax()
most_popular_club
```

 Q13: What percentage of students participated in the Coding Club?

```python
coding_club_pct = round(df['Clubs'].str.contains("Coding Club").mean() * 100, 2)
coding_club_pct
```

Q14: What is the most commonly mentioned skill among students?

```python
from collections import Counter
skills_list = df['Skills'].dropna().str.split(',').sum()
skills_cleaned = [skill.strip() for skill in skills_list]
skill_counts = Counter(skills_cleaned)
most_common_skill = skill_counts.most_common(1)[0]  # (Skill, Count)
print("Q14: Most Common Skill:", most_common_skill[0], "mentioned", most_common_skill[1], "times")
```

 Q15: What is the correlation between average GPA and CTC (LPA)?

```python
gpa_ctc_corr = round(df['Average GPA'].corr(df['CTC (LPA)']), 2)
gpa_ctc_corr
```

## **Data Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Placement Status Count
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Placement Status', palette='viridis')
plt.title('Placement Status Count')
plt.xlabel('Placement Status')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  CTC Distribution for Placed Students
plt.figure(figsize=(6, 4))
sns.histplot(data=df[df['CTC (LPA)'] > 0], x='CTC (LPA)', bins=20, kde=True, color='teal')
plt.title('CTC Distribution Among Placed Students')
plt.xlabel('CTC (LPA)')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

```python
# Plot 1: Distribution of Average GPA ===
plt.figure(figsize=(8, 5))
sns.histplot(df['Average GPA'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Average GPA')
plt.xlabel('Average GPA')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Plot 2: Count of Students by Branch ===
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Branch', palette='Set2')
plt.title('Number of Students by Branch')
plt.xlabel('Branch')
plt.ylabel('Count')
plt.show()
```

```python
#  Plot 3: Average GPA by Gender ===
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Gender', y='Average GPA', palette='pastel')
plt.title('Average GPA by Gender')
plt.xlabel('Gender')
plt.ylabel('Average GPA')
plt.show()
```

```python
placement_counts = df['Placement Status'].value_counts()

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    placement_counts,
    labels=placement_counts.index,
    autopct='%1.1f%%',
    colors=['#66b3ff', '#ff9999'],
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Placement Status of Students')
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```

```python
#  Plot 5: Correlation Heatmap ===
plt.figure(figsize=(12, 8))
correlation = df[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
#  Plot 6: Boxplot of CTC by Branch (only placed students) ===
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[df['Placement Status'] == 'Placed'], x='Branch', y='CTC (LPA)', palette='Set3')
plt.title('CTC Distribution by Branch (Placed Students Only)')
plt.xlabel('Branch')
plt.ylabel('CTC (LPA)')
plt.xticks(rotation=45)
plt.show()
```

```python
##violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Branch', y='Average GPA', palette='Set3')
plt.title('Distribution of Average GPA by Branch')
plt.xlabel('Branch')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
